# 프로젝트2 : Information Retrieval System

## 과제 내용

### [색인 & 가중치 => 학습]

1. 수집한 문서목록 가져오기
2. 각 문서에서 색인어 추출 전처리 -> 어절 -> 형태소 + 명사 + 바이그램
3. 가중치 계산(TF-IDF) : DTM -> TDM -> TWM (with document_vector_weight)

### [질의]

4. 질의에서 색인어 추출(색인과정의 2)
5. 가중치 계산(색인과정의 3)
6. 유사도 계산
7. 유사도 순 정렬(Euclidean distance:오름차순, Cosine Similarity:내림차순)
8. 검색결과 출력

### [kNN => k 선정]

9. k개의 가장 유사한 문서들의 카테고리 추출 (최근 뉴스 기사 1건을 무작위로 선정하여 query로 사용)
10. 카테고리 별 투표를 통해 가장 유사한 카테고리 선정 

### [스팸 분류]

11. 네이버 메일의 스펨 메일 scraping 하기

### [성능]

- Precision(정확율) : 검색결과는 최대한 많이
- Recall(재현율) : 유사도는 최대한 높이

### 질의에서 색인어 추출

```python
query = "서울시에 거래되는 아파트의 전세값은?"
```

```python
from collections import defaultdict

from konlpy.tag import Kkma

kkma = Kkma().morphs
query_repr = defaultdict(int)

for token in query.split():
    for morpheme in kkma(token):
        if len(morpheme) > 1:
            query_repr[morpheme] += 1
```

```python
query_repr
```

### 가중치 계산

```python
from functions.info_retrieval import max_tf, raw_idf

max_freq = max(query_repr.values())
query_weight = defaultdict(float)

for token, freq in query_repr.items():
    if token in global_lexicon_idf.keys():
        tf = max_tf(freq, max_freq, 0.5)
        idf = global_lexicon_idf[token]
        query_weight[token] = tf * idf
```

```python
query_weight
```

### 유사도 계산

#### Euclidean distance

```python
def euclidian(x, y):
    return (x-y) **2
```

```python
eucl_candidate_list = dict()

for index_term, posting_idx in global_lexicon.items():
    query_tfidf = 0
    
    if index_term in query_weight.keys():
        query_tfidf = query_weight[index_term]
    
    for filename, document_weight in twm[index_term].items():
        if filename not in eucl_candidate_list.keys():
            eucl_candidate_list[filename] = euclidian(query_tfidf, document_weight)
        else:
            eucl_candidate_list[filename] += euclidian(query_tfidf, document_weight)
```

```python
eucl_candidate_list
```

#### Cosine similarity

```python
def inner_product(x, y):    # x는 query_weight, y는 document_weight
    return x * y
```

```python
cosine_candidate_list = dict()

for index_term, q_weight in query_weight.items():
    if index_term in global_lexicon.keys():
        posting_idx = global_lexicon[index_term]
        
        while True:
            if posting_idx == -1:
                break
            
            posting_data = global_posting[posting_idx]
            posting_idx = posting_data[3]
            document_weight = posting_data[2]
            
            if global_document[posting_data[1]] not in cosine_candidate_list.keys():
                cosine_candidate_list[global_document[posting_data[1]]] = inner_product(q_weight, document_weight)
            else:
                cosine_candidate_list[global_document[posting_data[1]]] += inner_product(q_weight, document_weight)

for document_idx, sum_product in cosine_candidate_list.items():
    cosine_candidate_list[document_idx] /= global_document_weight[document_idx]
```

```python
cosine_candidate_list
```

### 유사도 순 정렬

```python
# euclidian distance : 오름차순 정렬
eucl_result_list = sorted(eucl_candidate_list.items(), key=lambda x:x[1])
```

```python
# Cosine similarity : 내림차순 정렬
cosine_result_list = sorted(cosine_candidate_list.items(), key=lambda x:x[1], reverse=True)
```

### 검색결과 출력

```python
# euclidian distance : 검색 결과
print("query: ", query)

for i, (document, distance) in enumerate(eucl_result_list[:3]):
    print("순위:{0} / 문서:{1} / 거리:{2}".format((i+1), document, distance))
    print("   document:{0}".format(collection[global_document.index(document)]))
```

```python
# Cosine similarity : 검색 결과
print("query: ", query)

for i, (document, distance) in enumerate(cosine_result_list[:3]):
    print("순위:{0} / 문서:{1} / 유사도:{2}".format((i+1), document, distance))
    print("   document:{0}".format(collection[global_document.index(document)]))
```

```python
from functions.info_retrieval import query_index
from functions.info_retrieval import eval_query_weight
from functions.info_retrieval import candidate_list_by_cosine
from functions.info_retrieval import cosine_sort
from functions.info_retrieval import result_print
```

```python
def ir(query, global_lexicon, global_posting, global_document, global_lexicon_idf, global_document_weight, collection, count=3):
    query_repr = query_index(query)
    query_weight = eval_query_weight(query_repr, global_lexicon_idf)
    candidate_list = candidate_list_by_cosine(query_weight, global_lexicon, global_posting, global_document, global_document_weight)
    result_list = cosine_sort(candidate_list)
    result_print(query, result_list, global_document, collection, count=count)
    return None
```
