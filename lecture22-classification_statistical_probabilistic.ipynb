## Classification Methods : Statistical/Probobilitic

- 분류, 추천에 적용 가능

- 개별 확률 : P(A), P(B)
- 조건부 확률 : P(A|B) = P(A,B) / P(B)
- 결합 확률 : P(A,B)

- $\Pr(A)$는 A의 사전 확률로, 아직 사건 B에 관한 어떠한 정보도 알지 못하는 것을 의미한다.
- $\Pr(A|B)$는 B의 값이 주어진 경우에 대한 A의 사후 확률이다.
- $\Pr(B|A)$는 A가 주어졌을 때 B의 조건부 확률이다.
- ${\mathcal  L}(A|B)=\Pr(B|A)$는 B가 주어졌을 때 A의 가능도이다.

### MLE(Maximum Likelihood Estimation) : Frequency Approch

- 확률 : 일정한 조건 아래에서 어떤 사건이나 사상(事象)이 일어날 가능성의 정도. 또는 그런 수치. <br>
   수학적으로는 1을 넘을 수 없고 음이 될 수도 없다. 확률 1은 항상 일어남을 의미하고, 확률 0은 절대로 일어나지 않음을 의미한다
- Binomial Distribution
    - probability : p
    - probability q = 1 - p
    - I.I.D condition : Independent events, Identically distributed random variables

$\normalsize P(H)=\theta, P(T)=1-\theta, (P\ge 0)$

$\normalsize P(D|\theta) = \theta^{\alpha_H}(1-\theta)^{\alpha_T}$  => 예) 5회 중에 앞면이 3회, 뒷면이 2회 나왔다면, $P(D|\theta) = \theta^{3}(1-\theta)^{2}$

$\normalsize \hat{\theta} = argmax_{\theta} P(D|\theta)$
$\normalsize = argmax_{\theta} \theta^{\alpha_H}(1-\theta)^{\alpha_T}$
$\normalsize = argmax_{\theta} \ln \theta^{\alpha_H}(1-\theta)^{\alpha_T}$
$\normalsize = argmax_{\theta} {\alpha_H} \ln \theta + {\alpha_T} (1-\theta)$

$\normalsize \frac{d}{d\theta} ({\alpha_H} \ln \theta + {\alpha_T} (1-\theta)) = 0$

$\normalsize \frac{\alpha_H}{\theta} - \frac{\alpha_T}{1-\theta} = 0$

$\normalsize \hat{\theta} = \frac{{\alpha_H}}{{\alpha_H} + {\alpha_T}}, N = {\alpha_H} + {\alpha_T}$

### MAP(Maximum A Posteriori Estimation)

베이지언 확률은 사후확률(posterior probability)을 사전확률(prior probability)과 <br>
likelihood를 이용해서 계산할 수 있도록 해 주는 확률 변환식이다. 

베이지안 확률은 영상처리에서 classification 문제, detection 문제가 나오면 거의 빠지지 않고 나오는 약방의 감초같은 존재로서, <br>
사람들이 자신의 방법론에 대해 무언가 수학적 모델이 필요할 경우에 1차적으로 생각하는게 바로 베이지언 확률이다.

☞ likelihood: $P(D|\theta)$, 어떤 모델에서 해당 데이터(관측값)이 나올 확률

☞ 사전확률(prior probability): $P(\theta)$, 관측자가 관측을 하기 전에 시스템 또는 모델에 대해 가지고 있는 선험적 확률.<br>

    예를 들어, 남여의 구성비를 나타내는 p(남자), p(여자) 등이 사전확률에 해당한다.
     
☞ 사후확률(posterior probability): $P(\theta|D)$, 사건이 발생한 후(관측이 진행된 후) 그 사건이 특정 모델에서 발생했을 확률 <br>

    => 확률 $P(\theta)$ given D => 사건 D가 발생 했을 때, $\theta$가 발생할 확률

#### Bayes’ Theorem

Bayes’ Theorem은 p(Y|X)와 p(X|Y)의 관계를 표현하는 식이다. 

식의 꼴은 매우 간단하지만, 이 theorem은 많은 의미를 가지고 있다. Thoerem은 다음과 같다.

$P(Y|X) = \large \frac{P(X|Y) P(Y)}{P(X)}$

이때 f(X|θ)를 likelihood, f(θ)를 prior, 그리고 f(θ|X)를 posterior라고 하며 <br>
각각은 observation(likelihood), 현상에 대한 사전정보 (prior), 주어진 데이터에 대한 현상의 확률 (posterior)을 의미한다.

이 식이 중요한 이유는 우리가 관측할 수 있는 데이터 이외에도 데이터에 대한 적절한 가정이 있다면 <br>
관측한 데이터만을 사용하는 것 보다 더 우수한 parameter estimation을 가능하게 하기 때문이다.

다시 MLE와 MAP로 돌아가보자. Bayes’ Theorem을 사용하면 MAP와 MLE의 관계를 다음과 같이 적을 수 있다.

$\hat{\theta} = argmax_{\theta} f(\theta|X) = argmax_{\theta} f(X|\theta) f(\theta) / f(X) = argmax_{\theta} \normalsize \frac{{\mathcal  L} (\theta|X) f(\theta)}{f(X)}$

이때, f(X) term은 θ 에 영향을 받는 term이 아니기 때문에 다음과 같이 적을 수 있다.

$\hat{\theta} \propto argmax_{\theta} {{\mathcal  L} (\theta|X) f(\theta)}$

#### Example

- **euclidian distance**

```python
from random import randint
```

```python
import matplotlib.pyplot as plt
from math import sqrt
```

```python
# D => 0:(1,1), 1:(2,2), ...
# C => 0:0, 1:1, ...
data = list()   # instance
category = list()

N = 20

for i in range(N):
    data.append((randint(1, 100), randint(1,100)))
    category.append(1 if sum(data[i]) > 100 else 0)
```

```python
def euclidian(x, y):
    return sqrt((x[0] - y[0]) ** 2 + (x[1] - y[1]) ** 2)
```

```python
x = [row[0] for row in data]
y = [row[1] for row in data]

similarity = list()

sample = (50, 50)

plt.rcParams["figure.figsize"] = (8, 8)
colorMap = ["b", "r"]

for i in range(N):
    similarity.append((i, euclidian(data[i], sample)))
    plt.scatter(x[i], y[i], facecolors="none", edgecolors=colorMap[category[i]])

plt.scatter(sample[0], sample[1], edgecolors="k")

plt.show()
```

```python
K = 3
candidateList = list()

for i, (idx, dist) in enumerate(sorted(similarity, key=lambda x:x[1])):
    if (i+1) > K:
        break
    candidateList.append(idx)    
```

```python
candidateCategory = list([0] * len(set(category)))

for idx in candidateList:
    candidateCategory[category[idx]] += 1
```

```python
colorMap = ["b", "r"]

for i in range(N):
    if i in candidateList:
        plt.scatter(x[i], y[i], facecolors="g", edgecolors=colorMap[category[i]])
    else:
        plt.scatter(x[i], y[i], facecolors="none", edgecolors=colorMap[category[i]])
    plt.text(x[i]+1, y[i]+2, "({0}:{1},{2}={3:.1f})".format(i, x[i], y[i], similarity[i][1]))

plt.scatter(sample[0], sample[1], facecolors=colorMap[candidateCategory.index(max(candidateCategory))], edgecolors="k")

plt.show()
```

- **cosine similarity**

```python
def cosine(x, y):
    innerProduct = 0
    xLength = 0.0
    yLength = 0.0
    
    for i in range(len(x)):
        innerProduct += x[i] * y[i]
    xLength = euclidian(x, (0,0))
    yLength = euclidian(y, (0,0))

    return innerProduct / (xLength * yLength)
```

```python
x = [row[0] for row in data]
y = [row[1] for row in data]

similarity = list()

sample = (50, 50)

plt.rcParams["figure.figsize"] = (8, 8)
colorMap = ["b", "r"]

for i in range(N):
    similarity.append((i, cosine(data[i], sample)))
    plt.scatter(x[i], y[i], facecolors="none", edgecolors=colorMap[category[i]])

plt.scatter(sample[0], sample[1], edgecolors="k")

plt.show()
```

```python
K = 3
candidateList = list()

for i, (idx, dist) in enumerate(sorted(similarity, key=lambda x:x[1], reverse=True)):   # 내림차순
    if (i+1) > K:
        break
    candidateList.append(idx)    
```

```python
candidateCategory = list([0] * len(set(category)))

for idx in candidateList:
    candidateCategory[category[idx]] += 1
```

```python
colorMap = ["b", "r"]

for i in range(N):
    if i in candidateList:
        plt.scatter(x[i], y[i], facecolors="g", edgecolors=colorMap[category[i]])
    else:
        plt.scatter(x[i], y[i], facecolors="none", edgecolors=colorMap[category[i]])
    plt.text(x[i]+1, y[i]+2, "({0}:{1},{2}={3:.1f})".format(i, x[i], y[i], similarity[i][1]))

plt.scatter(sample[0], sample[1], facecolors=colorMap[candidateCategory.index(max(candidateCategory))], edgecolors="k")

plt.show()
```
